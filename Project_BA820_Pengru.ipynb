##**Article Text - Pengru Lin**

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

cv_artext = CountVectorizer(stop_words = "english", strip_accents = "ascii")

cv_artext.fit(df["Article text"])

#print('number of `tokens`', len(cv_artext.vocabulary_))
#cv_artext.vocabulary_
```

*dataset is too big, use random 15% of the original dataset's to verify code*

```python
df_random_sample = df.sample(frac=0.15, random_state = 42)
```

```python
dtm_artext = cv_artext.transform(df_random_sample["Article text"])
bow_artext = pd.DataFrame(dtm_artext.toarray(), columns=cv_artext.get_feature_names_out())
bow_artext
```

```python
bow_artext.sum(axis = 0).sort_values(ascending = False)
```

```python
tfidf_artext_model = TfidfVectorizer(stop_words = "english", norm = "l2", strip_accents = "ascii")
tfidf_artext_model.fit(df_random_sample["Article text"])
```

```python
tfidf_transformed_artext = tfidf_artext_model.transform(df_random_sample["Article text"])
tfidf_artext = pd.DataFrame(tfidf_transformed_artext.toarray(), columns = tfidf_artext_model.get_feature_names_out())
tfidf_artext
```

```python
tfidf_artext[tfidf_artext["apple"] > 0]["apple"]
```

```python
tfidf_artext.sum(axis = 0).sort_values(ascending = False)
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt


word_freq = bow_artext.sum(axis=0).sort_values(ascending=False).to_dict()
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
